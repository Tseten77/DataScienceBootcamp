```python
import pandas as pd
data = pd.read_csv("iris.csv")
print(data)
data.groupby('variety').mean()
```

Assigning x to sepal.length, sepal.width, petal.length, and petal.width. Y to variety. After that, splitting x and y into test and train set.

```python
X=data.iloc[:,0:4]
Y=data.iloc[:,4]
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
import matplotlib.pyplot as plt
x_labels_list=['Training_set','Testing_set']
y_values_list=[len(X_train),len(Y_train)]
plt.bar(x_labels_list,y_values_list)
```
